```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
#avoid warnings
import warnings
# Suppress all warnings
warnings.filterwarnings("ignore")
# Suppress specific warning (e.g., FutureWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df=load_breast_cancer()
df
```

```python
# Independent Feature
X=pd.DataFrame(df['data'], columns=df['feature_names'])
X.head()
```

```python
#dependent feature
y=pd.DataFrame(df['target'], columns=['Target']) 
y.head()
```

```python
# Checking if the target is vbalanced or not
y['Target'].value_counts()
```

* this can be considered balanced

```python
# Train and test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

params=[{'C':[1,5,10]}, {'max_iter':[100,150]}]
model=LogisticRegression(C=100, max_iter=100)
model=GridSearchCV(model, param_grid=params, scoring='f1', cv=5)
model.fit(X_train, y_train)
```

```python
print(model.best_params_)
print(model.best_score_)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

### Interpretation:
#### True Negatives (TN): 59
* The model correctly predicted 0 for 59 instances.
#### False Positives (FP): 4
* The model incorrectly predicted 1 for 4 instances where the true label was 0.
#### False Negatives (FN): 1
* The model incorrectly predicted 0 for 1 instance where the true label was 1.
#### True Positives (TP): 107
* The model correctly predicted 1 for 107 instances.

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

### Key Metrics:
#### Precision:
* Measures the accuracy of positive predictions.
* Formula: **Precision**=TP / TP+FP

#### Recall (Sensitivity or True Positive Rate):
* Measures the ability to correctly identify all positive samples.
* Formula: **Recall** =TP / TP + FN
 
#### F1-Score:
* Harmonic mean of precision and recall.
* Formula: **F1-Score** = 2⋅Precision⋅Recall / Precision + Recall

#### Support:
* The number of actual occurrences of each class in the test set.
